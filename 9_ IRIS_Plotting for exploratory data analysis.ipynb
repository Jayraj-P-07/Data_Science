```python
# plotting a  pdf of petal length

counts, bin_edges=np.histogram(iris_setosa['petal.length'], bins=10, density=True)

pdf=counts/(sum(counts))
print(pdf)
print(counts)
print(bin_edges)


# plotting a pdf of petal length
cdf=np.cumsum(pdf)
plt.plot(bin_edges[1:], pdf)
plt.plot(bin_edges[1:],cdf)



```
