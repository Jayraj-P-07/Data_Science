## K-S Test for Similarity of Two Distributions

```python
# Importing packages
import numpy as np
import seaborn as sns
from scipy import stats
import matplotlib.pyplot as plt
```

```python
# Generating a random variable X (sample size =500) having Gaussian Distribution
x = stats.norm.rvs(size=500)
sns.set_style('whitegrid')
sns.kdeplot(np.array(x), bw_method=0.5)
plt.show()
```

```python
# Testing the generated distribution (X) belongs to normal distribution 
stats.kstest(x, 'norm')
```

p-value is found to be 0.9967 which is very high, so the distribution belongs to Normal distribution.

```python
# Generating a random variable Y (10000 sample size) having uniform distribution
y=np.random.uniform(0,1,10000)
sns.kdeplot(np.array(y), bw_method=0.1)
plt.show()
```

```python
# Testing the generated distribution (Y) belongs to normal distribution 
stats.kstest(y, 'norm')
```

p value is found to be 0.0 which is very low so the generated distribution doesn't belong to normal distribution
