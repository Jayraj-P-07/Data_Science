## Q-Q Plot

- These plots are useful to check whether the given distribution is Gaussian or not
- These are also useful to check two distributions's similarity

```python
import numpy as np
import pylab
import scipy.stats as stats
```

```python
# to generate 1000 numbers from Gaussian distribution having mean = 0 and variance =1
# X: N (0,1)
std_normal=np.random.normal(loc=0, scale=1, size= 1000)
```

```python
# to generate 1 to 100 percentiles of all values of X: N(0,1)

for i in range(0,101):
    print(i, np.percentile(std_normal, i))
```

```python
# to generate 100 samples from Y: N(20, 5), where 20 is the mean and 5 is the variance.
measurements=np.random.normal(20, 5, 100)
measurements
```

```python
# to plot 0 to 100 percentiles of Y: N(20, 5) with X: N(0,1)
# to check the Y: N(20,5) is gaussian or not

stats.probplot(measurements, dist='norm', plot=pylab)
pylab.show()
```

When we compare the percentiles of a given distribution i.e N(Mean, Variance) with standard normal distribution i.e. N(0, 1), if all the values lie on the straight line then its it Gaussian distribution else its not a Gaussian distribution.

## Here 100 percentiles of 1000 samples from standard normal distribution [X: N(0,1)] are taken and plotted against 100 percentiles of 100 sample taken

```python
# to generate 1000 numbers from Gaussian distribution having mean = 0 and variance =1
# X: N (0,1)
std_normal=np.random.normal(loc=0, scale=1, size= 1000)

# to generate 1 to 100 percentiles of all values of X: N(0,1)
X=[]
for i in range(0,101):
    print(i, np.percentile(std_normal, i))
    X.append(np.percentile(std_normal, i))
print(X)
```

```python
# to generate 100 numbers from Gaussian distribution having mean = 20 and variance =5
# Y: N (20,5)
normal=np.random.normal(loc=20, scale=5, size= 1000)

Y=[]
# to generate 1 to 100 percentiles of all values of Y: N(20, 5)
for i in range(0,101):
    print(i, np.percentile(normal, i))
    Y.append(np.percentile(normal, i))

print(Y)
```

```python
import matplotlib.pyplot as plt

plt.plot(X,Y, 'ro')
plt.xlabel("X: N(0, 1)")
plt.ylabel("Y: N(20, 5)")
plt.title("Plot between percentiles of X: N(0,1) and Y: N(20,5)")
plt.grid()
plt.show()
```
