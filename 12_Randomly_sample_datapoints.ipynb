```python
import random
print(random.random())
```

```python
# checking random number generation pattern for 10 times

for i in range(10):
    print("the {} random number is {} :".format(i, random.random()))
```

random.random() generates data between 0 and 1

```python
# Loading IRIS dataset from Sklearn
from sklearn import datasets
iris=datasets.load_iris()
df=iris.data
df.shape
```

```python
# checking dataframe
df
```

```python
# randomly sampling 30 points from 150 points using random number generator
n=150
m=30
p=m/n

sampled_data=[]

for i in range(0,n):
    if random.random() <= p:
        sampled_data.append(df[i,:])

print(len(sampled_data))
print(sampled_data)
```
