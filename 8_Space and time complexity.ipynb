```python
import math

def binarysearch(arr,l, r, x):
    
    if r>=l:
        mid=l+math.floor((r-l)/2)
        
        if arr[mid]==x:
            return mid
        
        elif arr[mid]>x:
            return binarysearch(arr,l,mid-1,x)
        
        elif arr[mid]<x:
            return binarysearch(arr,mid+1,r,x)
        
    else:
        return -1
    
lst=list(range(10))

binarysearch(lst,0,len(lst),1)
```

```python
l1=list(range(100))
l2=list(range(50))


random.shuffle(l1)
random.shuffle(l2)

count=0
for ele in l1:
    for item in l2:
        if l1[ele]==l2[item]:
            print(item)
            count=count+1
print("no. of common elements is {}".format(count))
        
```
