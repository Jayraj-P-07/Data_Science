```python
import matplotlib.pyplot as plt

plt.plot([1,2,3,4,5,6],[1,2,3,4,5,6])
plt.xlabel("indices")
plt.ylabel("numbers")
plt.title("example_plot")
plt.grid()
plt.show()
```

```python
plt.plot([1,2,3,4,5],[5,4,3,2,1], 'ro')
plt.xlabel("x Axis")
plt.ylabel("y Axis")
plt.title("my_plot")
plt.grid()
plt.show()
```

```python
import numpy as np
t=np.arange(0,5,0.2)
plt.plot(t, t**2, 'b--', label='^2')
plt.plot(t,t**2.2, 'rs', label='^2.2')
plt.plot(t,t**2.4, 'g^', label='2.4')
plt.grid()
plt.xlabel("x axis")
plt.ylabel('y axis')
plt.title("my second plot")
plt.legend()
plt.show()
```

```python
x=[2,3,4,6,6,8]
y=[3,5,6,3,5,6]

plt.plot(x,y,linewidth=3.0,)
plt.grid()
plt.show()
```

```python
# plotting two different lines

x=[2,3,4,6,6,8]
y=[3,5,6,3,5,6]

x1=[2,5,4,7,6,9]
y1=[3,2,9,2,5,4]

lines=plt.plot(x,y,x1,y1)

plt.setp(lines[0], color='g', linewidth=2.0)
plt.setp(lines[1], color='b', linewidth=1.0)

plt.grid()

plt.xlabel("x axis")
plt.ylabel("y axis")
plt.title("jerry's plot")

plt.show()
```

```python
# multiple plots
def f(t):
    return np.exp(t)

t1=np.arange(1,10,2)
t2=np.arange(1,20,2)

plt.figure(1)

plt.subplot(211)
plt.grid()
plt.plot(t1, f(t1), 'b--')

plt.subplot(212)
plt.grid()
plt.plot(t2,f(t2), 'ro')
plt.show()
```
